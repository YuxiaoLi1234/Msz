```python
import matplotlib.pyplot as plt
import numpy as np

# 
threads = [1, 2, 4, 8, 16, 32, 64, 128]
times = [0.0215886, 0.0627011, 0.00786238, 0.00663097, 0.00717736, 0.0488605, 0.0100836, 0.0681726]

# 
log_threads = np.log2(threads)
log_times = np.log10(times)

# 
ideal_times = times[0] / np.array(threads)

# 
plt.figure(figsize=(8, 6))
plt.plot(threads, times, 'o-', label='Actual Timing', color='gold')
plt.plot(threads, ideal_times, '--', label='Ideal Scaling', color='skyblue')

# 
percentages = [100 * t / times[0] for t in times]
for i, (x, y, p) in enumerate(zip(threads, times, percentages)):
    plt.text(x, y, f'{int(p)}%', fontsize=10, ha='right' if i == 0 else 'left')

# 
plt.xscale('log', base=2)
plt.yscale('log', base=10)

# 
plt.xlabel('Number of Threads')
plt.ylabel('Time')
plt.title('(a)')
plt.legend()
plt.grid(True, which="both", ls="--")

# 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# : 
data = {
    1: [75312],
    2: [53667.7, 45814.4],
    4: [47190.8, 38501.3, 33183.2],
    8: [29565.9, 26639.1, 26227],
    16: [22224.9, 19369.7],
    32: [34104.7, 20505.9],
    64: [28761.7],
    128: [28924.9]
}
x = [1,2,4,8,16,32,64,128]
# 
threads = sorted(data.keys())
times = [np.mean(data[t]) for t in threads]

# 
log_threads = np.log2(threads)
log_times = np.log10(times)

# 
ideal_times = times[0] / np.array(threads)

# 
plt.figure(figsize=(10, 6))
plt.plot(threads, times, 'o-', label='Actual Timing', color='gold')
plt.plot(threads, ideal_times, '--', label='Ideal Scaling', color='skyblue')

# 
percentages = [100 * t / times[0] for t in times]
for i, (x, y, p) in enumerate(zip(threads, times, percentages)):
    plt.text(x, y, f'{int(p)}%\n{y/1000:.2f}s', fontsize=10, ha='center')

# 
plt.xscale('log', base=2)
plt.yscale('log', base=10)
plt.xticks(threads, [str(t) for t in threads])
# 
plt.xlabel('Number of Threads', fontsize=14)
plt.ylabel('Time (ms)', fontsize=14)
plt.title('(d)', fontsize=16)
plt.legend()
plt.grid(True, which="both", ls="--")

# 
plt.show()
```
